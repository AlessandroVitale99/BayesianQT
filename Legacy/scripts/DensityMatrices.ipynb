```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
from qiskit.visualization import array_to_latex
```

```python
ψ0 = qi.Statevector([0, 1, 0, 0])
ψ0.draw('latex')
```

```python
qc = QuantumCircuit(2)
qc.h(1)
qc.cx(1,0)
qc.draw(reverse_bits=True)
```

```python
ψ1 = ψ0.evolve(qc)
ψ1.draw('latex')
```

```python
ρ1 = qi.DensityMatrix(ψ1)
ρ1.draw('latex')
```

```python
ρ1.draw('city')
```

## Evolution of a pure-state density operator

```python
ρ2 = qi.DensityMatrix.from_label('l')
ρ2.draw('latex')
```

```python
S = qi.Operator.from_label('S')
array_to_latex(S)
```

```python
ρ3 = ρ2.evolve(S)
ρ3.draw('latex')
```

## Evolution of a mixed-state density operator

```python
ρ4 = 1/3 * qi.DensityMatrix.from_label('0') + 2/3 * qi.DensityMatrix.from_label('1')
ρ4.draw('latex')
```

```python
ρ4.evolve(qi.Operator.from_label('X')).draw('latex')
```
